```python
# !pip install keras
import keras
from keras.preprocessing import image

from keras.utils import load_img, img_to_array
import os
# from keras.preprocessing.image import ImageDataGenerator 


datagen = image.ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip = True,
    fill_mode='nearest'
)
folder = '/content/drive/MyDrive/Dataset/BaseDataset/'
for filename in os.listdir(folder):
    img =load_img(folder+filename )
    x=img_to_array(img)
    x=x.reshape((1,)+x.shape)

    i=0
    for batch in datagen.flow(x, batch_size=1,
                              save_to_dir='/content/drive/MyDrive/Dataset/img', save_prefix=filename, save_format='jpeg'):
        i += 1
        if i > 70:
            break 
```

```python
import shutil
import random
import math
import os
import numpy as np

root_dir = '/content/drive/MyDrive/Dataset/img'
root_label_dir = '/content/drive/MyDrive/Dataset/labels'

files = os.listdir(root_dir)
labels = os.listdir(root_label_dir)
np.random.shuffle(files)

base_img_outdir = '/content/drive/MyDrive/Dataset/images/'
base_label_outdir ='/content/drive/MyDrive/Dataset/label/'

for folder in ['train', 'val']:
    target_img_dir = base_img_outdir + folder
    target_label_dir = base_label_outdir + folder

    if folder == 'train':
        images_to_pass = files[: math.floor(0.8*len(files))]
        for img in images_to_pass:
            x = img.split('.jpeg')[0]
            img = root_dir + '/' + img
            
            lab = root_label_dir +'/'+ x+'.txt'
            print(img)
            print(lab)
            shutil.copy(img, target_img_dir)
            shutil.copy(lab, target_label_dir)
    else :
        images_to_pass = files[math.floor(0.8*len(files)): math.floor(0.9*len(files))]
        for img in images_to_pass:
            x = img.split('.jpeg')[0]
            img = root_dir + '/' + img
            
            lab = root_label_dir +'/'+ x+'.txt'
            print(img)
            print(lab)
            shutil.copy(img, target_img_dir)
            shutil.copy(lab,target_label_dir)
    
```
